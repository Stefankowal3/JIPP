### Zbiory

```python
# Utwórz zbiór o nazwie "flowers" składający się z 3 elementów
flowers = {'lily','rose','lotus'}
```

```python
flowers
```

```python
# Dodaj element "jasmine" do zbioru "flowers"
flowers.add('jasmine')
```

```python
# Policz liczbę elementów w zbiorze "flowers"
len(flowers)
```

```python
# Dodaj elementy za zbioru "other" do zbioru "flowers"
other = {'tulip'}
flowers.update(other)
```

```python
flowers
```

```python
# Usuń "jasmine" ze zbioru "flowers"
flowers.remove('jasmine')
```

```python
flowers
```

Element "jasmine" został usunięty ze zbioru

```python
# Ponowne usunięcie elementu "jasmine" za pomocą metody remove() zwróci błąd
flowers.remove('jasmine')
```

```python
# Spróbujmy usunąć "jasmine" za pomocą discard()
flowers.discard('jasmine')
```

Jeśli w danym zbiorze wskazany przez użytkownika element do usunięcia nie istnieje, metoda remove() zwróci błąd, w przeciwieństwie do metoda discard()

```python
# Metoda pop() używamy do usunięcia ostatniego elementu ze zbioru
x = flowers.pop()
```

Uwaga! Zbiory nie są uporządkowane, więc wywołując metodę pop() nie wiemy, który element zostanie usunięty

### Program do rezerwacji filmów

W tym programie będziemy używać
* struktur danych takich jak słownik, lista
* Instrukcji warunkowych if-else
* Pętli while
* Rzutowanie (konwersja) typów 



```python
# Utwórz słownik filmów. Niech  kluczem będzie nazwa filmu, a parą wartości dwie liczby: kryteria wiekowe oraz liczba dostępnych biletów

movies = {"Finding Nemo":[5,2], "Moana":[6,3], "Batman":[18,5], "The Lion King":[10,4]}

# Utwórz pętlę, która będzie działać w nieskończoność

while True:
    
    # Pobierz tytuł filmu od użytkownika, usuń spacje z początku i końca a następnie zamień frazę na format tytułowy (pierwsza litera każdego słowa jest wielka)
    movie_title = input("Enter movie title: ").strip().title()
    
    # Stwórz instrukcję warunkową if. Jeśli wybrany film jest odostępny w słowniku, kontynuuj
    if movie_title in movies:
        
        # Zapytaj użytkownika o wiek
        age = int(input("Enter your age: "))
        
        # Sprawdź użytkownika pod kątem kwalifikowalności
        if age >= movies[movie_title][0]:
            
            # Jeśli użytkownik znajduje się w grupie docelowej, sprawdź dostępność miejsc
            if movies[movie_title][1] > 0:
                
                # Jeśli liczba dostępnych miejsc jest wartością dodatnią, zmiejsz pulę dostępnych miejsc o 1
                movies[movie_title][1] -= 1
                print(f"Ticket booked! Available tickets: {movies[movie_title][1]}")
            else:
                print("Sorry, no tickets available!")
        else:
            print(f"Sorry, you are too young. This movie is for ages {movies[movie_title][0]}+")
    else:
        print("Movie not found!")
            
```

## Moduł random

```python
# Moduł do losowania liczb
import random
```

```python
# randrange() - Ta metoda zwraca losową liczbę z określonego zakresu
# Składnia: random.randrange(start, stop, krok)

# Zwraca liczbę między 10 (łącznie) a 26 (rozłącznie). Krok wynosi 1, więc zwróci wszystkie liczby z tego zakresu. Jeśli krok jest większy niż 1, zwróci tylko liczby, które są wielokrotnościami tego kroku.
random.randrange(10,26,1)
```

```python
# randint() - Ta metoda zwraca liczbę całkowitą z określonego zakresu
# Składnia: random.randint(start, stop)

# Zwraca liczbę od 50 do 100 (obie wartości są włączone)
random.randint(50,100)
```

```python
# choice() - Ta metoda zwraca losowy element z określonej sekwencji
# Składnia: random.choice(sekwencja)

x = ['Python','Program','Language']
random.choice(x)
```
